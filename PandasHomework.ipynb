```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data-frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
#the total number of players 
player= len(purchase_data["SN"].value_counts())
player_count=pd.DataFrame([player], columns = ["Total Players"])
player_count
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
average_price = purchase_data['Price'].mean()
unique_items = purchase_data['Item ID'].nunique()

number_purchases = len(purchase_data)
total_revenue = purchase_data['Price'].sum()



summary_table = pd.DataFrame({"Number of Unique Items": [unique_items],
                              "Average Price": [average_price],
                              "Number of Purchases": [number_purchases],
                              "Total Revenue": [total_revenue]})
summary_table["Average Price"] = summary_table["Average Price"].astype(float).map("${:,.2f}".format)
summary_table["Total Revenue"] = summary_table["Total Revenue"].astype(float).map("${:,.2f}".format)

summary_table
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
unique_dataframe = purchase_data.groupby("SN").first()


unique_dataframe["Gender"].value_counts()

unique_gender = unique_dataframe["Gender"].value_counts()
female_count = unique_dataframe["Gender"].value_counts()["Female"]
male_count = unique_dataframe["Gender"].value_counts()["Male"]


other_count = unique_dataframe["Gender"].value_counts()["Other / Non-Disclosed"]
total_count = female_count + male_count + other_count 

gender_list = unique_dataframe["Gender"].unique()
type_gender = pd.Series((v[0] for v in gender_list))


counter = (female_count, male_count, other_count)
percentage_female = female_count / total_count * 100
percentage_male = male_count / total_count * 100
percentage_other = other_count / total_count * 100

percentages = (percentage_male, percentage_female, percentage_other)
gender_table = pd.DataFrame({"Gender": type_gender,
                            "Total Count": counter,
                            "Percentage of Players": percentages})

gender_table["Percentage of Players"] = gender_table["Percentage of Players"].astype(float).map("{:.2f}%".format)
gender_table
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
gender_df = purchase_data.groupby(["Gender"])

purchase_count = gender_df["Price"].count()
price_mean = gender_df["Price"].mean()
total_purchase = gender_df["Price"].sum()



gender_purchasing = pd.DataFrame({"Purchase Count": purchase_count,
                                  "Average Purchase Price": price_mean,
                                  "Total Purchase Value": total_purchase,})

gender_purchasing["Average Purchase Price"] = gender_purchasing["Average Purchase Price"].astype(float).map("${:,.2f}".format)
gender_purchasing["Total Purchase Value"] = gender_purchasing["Total Purchase Value"].astype(float).map("${:,.2f}".format)

gender_purchasing
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bin_age = [0, 9, 14, 19, 24, 29, 34, 39, 1000]

group_names_age = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]



purchase_data["Age Summary"] = pd.cut(purchase_data["Age"], bin_age, labels= group_names_age)
age_df = purchase_data.groupby(["Age Summary"])

percentage_players_age = age_df["SN"].nunique() / total_count * 100
count_players_age = age_df["SN"].nunique()

age_summary_table = pd.DataFrame({"Total Count": count_players_age,
                                  "Percentage of Players": percentage_players_age})

age_summary_table["Percentage of Players"] = age_summary_table["Percentage of Players"].astype(float).map("{:.2f}".format)
age_summary_table
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
count_purchase_age = age_df["SN"].count()
avg_price_age = age_df["Price"].sum() / count_purchase_age

purchase_value = avg_price_age * count_purchase_age
avg_purchase_person = age_df["SN"].nunique() / age_df["Price"].sum()

age_purchase_table = pd.DataFrame({"Purchase Count": count_purchase_age,
                                  "Total Purchase Value": purchase_value,
                                  "Average Purchase Price": avg_price_age,
                                  "Avg Total Purchase per Person": avg_purchase_person})

age_purchase_table["Total Purchase Value"] = age_purchase_table["Total Purchase Value"].astype(float).map("${:,.2f}".format)
age_purchase_table["Average Purchase Price"] = age_purchase_table["Average Purchase Price"].astype(float).map("${:,.2f}".format)
age_purchase_table["Avg Total Purchase per Person"] = age_purchase_table["Avg Total Purchase per Person"].astype(float).map("${:,.2f}".format)

age_purchase_table
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
spenders_df = purchase_data.groupby(["SN"])

avg_price_age = age_df["Price"].sum() / count_purchase_age
purchase_value = count_purchase_age * avg_price_age 
avg_purchase_person = age_df["SN"].nunique() / age_df["Price"].sum() 


purchase_count_spender = spenders_df["SN"].count()
avg_purchase_price_spender = spenders_df["Price"].sum() / spenders_df["SN"].count()
total_purchase_price_spender = avg_purchase_price_spender * purchase_count_spender


spender_summary_table = pd.DataFrame({"Purchase Count": purchase_count_spender,
                                     "Total Purchase Value": total_purchase_price_spender,
                                      "Average Purchase Price": avg_purchase_price_spender,})

spender_summary_table_sort = spender_summary_table.sort_values("Purchase Count", ascending=False)

spender_summary_table_sort["Total Purchase Value"] = spender_summary_table_sort["Total Purchase Value"].astype(float).map("${:,.2f}".format)
spender_summary_table_sort["Average Purchase Price"] = spender_summary_table_sort["Average Purchase Price"].astype(float).map("${:,.2f}".format)


spender_summary_table_sort.head(5)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
items_df = purchase_data.groupby(["Item ID"])

item_name = items_df["Item Name"].max()
purchase_count_items = items_df["SN"].count()
price_items = items_df["Price"].max()
total_purchase_price_items = price_items * purchase_count_items



items_summary_table = pd.DataFrame({"Item Name": item_name,
                                    "Purchase Count": purchase_count_items,
                                     "Item Price": price_items,
                                     "Total Purchase Value": total_purchase_price_items})

items_summary_table_sort = items_summary_table.sort_values("Purchase Count", ascending=False)

items_summary_table_sort["Item Price"] = items_summary_table_sort["Item Price"].astype(float).map("${:,.2f}".format)
items_summary_table_sort["Total Purchase Value"] = items_summary_table_sort["Total Purchase Value"].astype(float).map("${:,.2f}".format)

items_summary_table_sort.head(5)
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
items_summary_table_new_sort = items_summary_table.sort_values("Total Purchase Value", ascending=False)

items_summary_table_new_sort["Item Price"] = items_summary_table_new_sort["Item Price"].astype(float).map("${:,.2f}".format)
items_summary_table_new_sort["Total Purchase Value"] = items_summary_table_new_sort["Total Purchase Value"].astype(float).map("${:,.2f}".format)

items_summary_table_new_sort.head(5)
```
